```python
import pandas as pd
```

```python
df = pd.read_excel("Takasbank TEFAS  Tarihsel Veriler.xlsx",skiprows=1)
```

```python
pivot = df.pivot_table(columns="Tarih",index=["Fon Kodu","Fon Adı"],values="Fiyat")
```

```python
pivot.fillna(0,inplace=True)
```

```python
pivot.drop(pivot[pivot.iloc[:,0]==0].index,inplace=True)
pivot.drop(pivot[pivot.iloc[:,-2]==0].index,inplace=True)
```

```python
pivot["getiri"] = (pivot.iloc[:,-1] - pivot.iloc[:,0])/pivot.iloc[:,0]*100
```

```python
pivot.sort_values("getiri",ascending=False)[[pd.Timestamp('2023-12-06'),pd.Timestamp('2024-01-26'),"getiri"]][:20]
```
