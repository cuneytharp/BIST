```python
import pandas as pd 
from tvDatafeed import TvDatafeed,Interval,TvDatafeedLive
import seaborn as sns
```

```python
tv = TvDatafeed()
```

```python
total_df = pd.DataFrame()
coins = "ALICEUSDT,KAVAUSDT,MANAUSDT,SUSHIUSDT,CAKEUSDT,MATICUSDT,SOLUSDT,AVAXUSDT,XLMUSDT,ADAUSDT,XRPUSDT,DOTUSDT,BNBUSDT,ETHUSDT,BTCUSDT"
coin_list = [x for x in coins.split(",")]
coin_list
```

```python
for coin in coin_list:
    data = tv.get_hist(coin,"BINANCE",interval=Interval.in_5_minute,n_bars=5000)
    data.columns= [coin,f"{coin}OPEN",f"{coin}HIGH",f"{coin}LOW",f"{coin}CLOSE","f{coin}VOL"]
    new_data = data[f"{coin}CLOSE"]
    total_df = pd.concat([total_df,new_data],axis=1)
```

```python
total_df
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(10,10))
sns.heatmap(total_df.corr(method="pearson"),annot=True,fmt="0.2f")
```
